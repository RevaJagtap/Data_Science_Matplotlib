# Hist plot

```python
import numpy as np
```

### 3 histogram in one plot

```python
plt.hist(a1)
plt.hist(a2)
plt.hist(a3)
 
```

# pie plot

 - univariate plot
 - used to display %wise distrbustion of data
 - used for Category-wise  comparison

```python
team = ['Ganesh','Seema','Abhishek']
captial = [12,10,20] #Cr.
```

```python
plt.pie(captial)
plt.show()
```

```python
team
```

```python
# display name of each teammate on pie
plt.pie(captial,labels=team)
plt.show()
```

```python
# add % on pie
plt.pie(captial,labels=team,autopct='%.f%%')
#plt.pie(captial,labels=team,autopct='%.2f%%')
#plt.pie(captial,labels=team,autopct='%f%%')
plt.show()
```

```python
# highlight /separate pie for Seema
plt.pie(captial,labels=team,autopct='%.2f%%',explode=[0.5,0.2,0],
       pctdistance=1.3,labeldistance=.4)
plt.show()
```

# subplot in Python

- when we want to combine multiple plots in a single canvas
- so we can have lets say 2 plots with difffent axis and its existance then use subplot option

```python
display(team,captial)
```

```python
plt.subplot(121)#1(row) 2(columns) 1(plot)
plt.bar(team,captial)
plt.subplot(122)
plt.plot(team,captial)
```

```python
plt.subplot(211)#1(row) 2(columns) 1(plot)
plt.bar(team,captial)
plt.subplot(212)
plt.plot(team,captial)
```

```python
# 2 x 2 subplot
plt.subplot(221)
plt.subplot(222)
plt.subplot(223)
plt.subplot(224)
```

```python
plt.subplot(221)
plt.bar(team,captial)

plt.subplot(222)
plt.pie(captial,autopct='%.f%%',labels=team)

plt.subplot(223)
plt.scatter(team,captial)

plt.subplot(224)
plt.plot(a1,color='r')
```

```python
#to add more margin betwen images use tight_layout
plt.subplot(221)
plt.bar(team,captial)

plt.subplot(222)
plt.pie(captial,autopct='%.f%%',labels=team)

plt.subplot(223)
plt.scatter(team,captial)

plt.subplot(224)
plt.plot(a1,color='r')
plt.tight_layout()
```
