# Line plot

```python
# Day vs Temperature
day = [1,2,3,4,5,6,7]
temp = [15,16,18,10,19,20,6]
```

```python
import matplotlib.pyplot as plt
```

```python
plt
```

```python
import matplotlib
matplotlib.__version__
```

```python
#line plot: we can use to show the progrss of data, to show the performace, 
# contineous changes in data,check flow of something
# plt.plot(input,output)
plt.plot(day,temp)
```

```python
# change a color
plt.plot(day,temp,color='black')
# white black yellow cyan magenta green red 
```

```python
# change a color
plt.plot(day,temp,color='red')
# white black yellow cyan magenta green red 
```

```python
#use some colors with its hex value from https://html-color.codes/
plt.plot(day,temp,color='#76ccf3')
```

```python
# add linestyle
plt.plot(day,temp,color='#76ccf3',linestyle='--',linewidth=6)
```

```python
# add linestyle
plt.plot(day,temp,color='#76ccf3',linestyle='--',linewidth=3)
```

```python
# use numpy array for random number generation and do the line plot
import numpy as np
a = np.random.random(100)
a
```

```python
# Line plot is Univariate, Bivariate,Multivariate
plt.plot(a)
```

```python
# add labels to plot
plt.plot(day,temp,color='#76ccf3',linestyle='--',linewidth=6)
plt.xlabel('Days',fontsize=12)
plt.ylabel('Temperature',fontsize=12)
plt.title('Weekly data of Temp')
plt.grid()
```

```python
#multivariate plot
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 300

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
# to mark intersection point
# use marker attribute
# add labels to plot
plt.plot(day,temp,color='red',linestyle='',linewidth=6,marker='o')
plt.xlabel('Days',fontsize=12)
plt.ylabel('Temperature',fontsize=12)
plt.title('Weekly data of Temp')
plt.grid()
```

```python
# add a plot label
plt.plot(day,temp,color='red',linestyle='',linewidth=6,marker='o',label='Daywise Temp')
plt.xlabel('Days',fontsize=12)
plt.ylabel('Temperature',fontsize=12)
plt.title('Weekly data of Temp')
plt.grid()
plt.legend(loc=8)
```
