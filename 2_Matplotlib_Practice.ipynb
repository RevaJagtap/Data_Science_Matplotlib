```python
import matplotlib.pyplot as plt
```

```python
values = [12,34,22,33,45,65,45,21,21,22,34,32,31]
```

```python
#univariate: only single input is used
# index is on X axis
# values on y axis
plt.plot(values,marker='*',markersize=10)
```

```python
plt.plot(values,marker='*',markersize=10,color='red',linestyle='--')
plt.xlabel('Data')
plt.ylabel('Count')
plt.title('Simple Graph')
```

```python
# Q. how to save the image of matplotlib
plt.plot(values,marker='*',markersize=10,color='red',linestyle='--')
plt.xlabel('Data')
plt.ylabel('Count')
plt.title('Simple Graph')
plt.savefig('sample.png',dpi=300)
#dpi: dots per inch
```

# Scatter plot

- it is used to display distribustion of data in terms of circles/markers
- It is completely Bivariate plot
- it is used to check outliers in data
- outliers are an exceptional cases among data present
- a data point far away from the distrubustion is an outlier

```python
import numpy as np
y = np.random.random(45)
y
```

```python
x = range(45)
x
```

```python
plt.scatter(x,y)
```

```python
#use different markers
plt.scatter(x,y,marker='D')
```

```python
# PS: plot overs vs runs
#take 10 overs data
overs = range(1,11)
runs = [0,2,10,4,16,20,1,4,6,8]
```

```python
# we want runs in cummulative order
runs= np.cumsum(runs)
runs
```

```python
#plt.scatter(overs,runs,color='k')
plt.plot(overs,runs,color='k') # line
plt.plot(overs,runs,color='cyan',marker='*',linestyle=' ',markersize=10) # markers
```

```python
# fall of wickets
f_w = [None,  2, None, None, 32, None, None, None, None, 71]
```

```python
plt.plot(overs,runs,label='runs per over')
plt.scatter(overs,f_w,color='red',label='wickets')
plt.xlim((1,9))
plt.ylim([0,71])
plt.legend()
```
